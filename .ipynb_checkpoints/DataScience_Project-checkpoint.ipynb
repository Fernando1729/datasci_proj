```r
gas_series.train = dataframe[which((dataframe$Product == "GASOLINA COMUM") & (dataframe$Year) < 2019),7]
gas_series.test = dataframe[which((dataframe$Product == "GASOLINA COMUM") & (dataframe$Year) > 2018),7]
```

```r
plot(gas_series.test)
plot(gas_series.train)
```

```r
dwnn <- function(query, dataset, sigma) {
    if (!is.matrix(dataset)) {
        return ("dataset must be a data frame.")
    }
    classId = ncol(dataset)
    # Conjunto de entrada
    X = matrix(dataset[,1:(classId-1)], ncol=classId-1)
    # Conjunto de saída
    Y = dataset[,classId]
    E = apply(X, 1, function(row) { sqrt(sum((row - query)^2)) } )
    act = exp(-E^2 / (2*sigma^2))
    return ( act%*%Y / sum(act) )
}

mypredict <- function(train.set,test.set,m, d,sigma,plot=F,decomposition=F,pca=F) {
    
    if (decomposition) {
        cat("Finding the deterministic component...\n")
        res.train = EMD::emd(train.set, boundary="wave")
        res.test = EMD::emd(test.set, boundary="wave")
        deterministic.train = rowSums(res.train$imf[,4:res.train$nimf]) + res.train$residue
        deterministic.test = rowSums(res.test$imf[,4:res.test$nimf]) + res.test$residue
        cat("Reconstructing the space phase....\n")
        dataset.train = tseriesChaos::embedd(deterministic.train , m=m, d=d)
        dataset.test = tseriesChaos::embedd(deterministic.test , m=m, d=d)
    } else {
        cat("Reconstructing the space phase....\n")
        dataset.train = tseriesChaos::embedd(train.set, m=m, d=d)
        dataset.test = tseriesChaos::embedd(test.set, m=m, d=d)
    }
    if (pca) {
        dataset.train.pca = prcomp(dataset.train[,1:ncol(dataset.train)-1], center = TRUE, scale = TRUE)
        print((summary(dataset.train.pca)))
    }
    
    #plot(dataset)
    #last = floor(nrow(dataset)*train.set)
    #cat("Separating train and test sets....\n")
    #train = dataset[1:last,]
    #test = dataset[(last+1):nrow(dataset),]
    
    buffer = dataset.train
    cat("Predicting....\n")
    cat(nrow(buffer),ncol(buffer),"\n")
    cat(nrow(dataset.test),ncol(dataset.test),"\n")
    last = nrow(dataset.train)
    for (i in (last+1):nrow(dataset.test)) {
        query = as.numeric(buffer[i - d, 2:ncol(buffer)])
        y = as.numeric(dwnn(query, dataset.train, sigma))
        buffer = rbind(buffer, c(query, y))
    }

    cat("Ploting....\n")
    if (plot) {
        plot(dataset.test, t="l")
        lines(buffer[nrow(dataset.train):nrow(dataset.test),ncol(dataset.test)], col=2) # aproximação
    }

    cat("Error...\n")
    error = dtw::dtw(dataset.test,buffer[(last+1):(last+1+100),ncol(dataset.train)])$normalizedDistance
    cat(error,"\n")

    cat("End\n")
    ret = list()
    ret$last = last
    ret$train = dataset.train
    ret$buffer = buffer

    return (ret)
}
```

```r
mypredict(train.set = gas_series.train[1:1000],test.set = gas_series.test,m = 5,d = 100,sigma = 0.054,plot = T)
```
